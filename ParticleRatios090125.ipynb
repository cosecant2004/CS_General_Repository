```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
```

```python
filepath_200GeV = "/Users/christianschmidt/Desktop/trajectum-2.1/src/200GeV_FinalResult1_NoSmash_80000_500PJ.h5"    
filepath_27GeV = "/Users/christianschmidt/Desktop/trajectum-2.1/src/27GeV_FinalResult1_NoSmash_80000_400PJ.h5"
filepath_19GeV = "/Users/christianschmidt/Desktop/trajectum-2.1/src/19GeV_FinalResult1_NoSmash_80000_400PJ.h5"
filepath_7GeV = "/Users/christianschmidt/Desktop/trajectum-2.1/src/7GeV_FinalResult1_NoSmash_80000_200PJ.h5"
```

```python
centrality_dir = "centrality"

multiplicitychargedpions_vals_dir = "multiplicitychargedpion/STARTPC2/centralitybinned/values"
multiplicitychargedpions_uerrs_dir = "multiplicitychargedpion/STARTPC2/centralitybinned/uppererrors"
multiplicitychargedpions_lerrs_dir = "multiplicitychargedpion/STARTPC2/centralitybinned/lowererrors"

multiplicitychargedkaons_vals_dir = "multiplicitychargedkaon/STARTPC2/centralitybinned/values"
multiplicitychargedkaons_uerrs_dir = "multiplicitychargedkaon/STARTPC2/centralitybinned/uppererrors"
multiplicitychargedkaons_lerrs_dir = "multiplicitychargedkaon/STARTPC2/centralitybinned/lowererrors"

multiplicityprotons_vals_dir = "multiplicityproton/STARTPC2/centralitybinned/values"
multiplicityprotons_uerrs_dir = "multiplicityproton/STARTPC2/centralitybinned/uppererrors"
multiplicityprotons_lerrs_dir = "multiplicityproton/STARTPC2/centralitybinned/lowererrors"
```

```python
with h5py.File(filepath_27GeV, "r") as hdf:
    multiplicitychargedpions_27GeV_vals = hdf[multiplicitychargedpions_vals_dir][:]
    multiplicitychargedpions_27GeV_uerrs=hdf[multiplicitychargedpions_uerrs_dir][:]
    multiplicitychargedpions_27GeV_lerrs=hdf[multiplicitychargedpions_lerrs_dir][:]

    multiplicitychargedkaons_27GeV_vals = hdf[multiplicitychargedkaons_vals_dir][:]
    multiplicitychargedkaons_27GeV_uerrs=hdf[multiplicitychargedkaons_uerrs_dir][:]
    multiplicitychargedkaons_27GeV_lerrs=hdf[multiplicitychargedkaons_lerrs_dir][:]

    multiplicityprotons_27GeV_vals = hdf[multiplicityprotons_vals_dir][:]
    multiplicityprotons_27GeV_uerrs=hdf[multiplicityprotons_uerrs_dir][:]
    multiplicityprotons_27GeV_lerrs=hdf[multiplicityprotons_lerrs_dir][:]
    centrality = hdf[centrality_dir][:]

with h5py.File(filepath_19GeV, "r") as hdf:
    multiplicitychargedpions_19GeV_vals = hdf[multiplicitychargedpions_vals_dir][:]
    multiplicitychargedpions_19GeV_uerrs=hdf[multiplicitychargedpions_uerrs_dir][:]
    multiplicitychargedpions_19GeV_lerrs=hdf[multiplicitychargedpions_lerrs_dir][:]

    multiplicitychargedkaons_19GeV_vals = hdf[multiplicitychargedkaons_vals_dir][:]
    multiplicitychargedkaons_19GeV_uerrs=hdf[multiplicitychargedkaons_uerrs_dir][:]
    multiplicitychargedkaons_19GeV_lerrs=hdf[multiplicitychargedkaons_lerrs_dir][:]

    multiplicityprotons_19GeV_vals = hdf[multiplicityprotons_vals_dir][:]
    multiplicityprotons_19GeV_uerrs=hdf[multiplicityprotons_uerrs_dir][:]
    multiplicityprotons_19GeV_lerrs=hdf[multiplicityprotons_lerrs_dir][:]
    centrality = hdf[centrality_dir][:]

with h5py.File(filepath_7GeV, "r") as hdf:
    multiplicitychargedpions_7GeV_vals = hdf[multiplicitychargedpions_vals_dir][:]
    multiplicitychargedpions_7GeV_uerrs=hdf[multiplicitychargedpions_uerrs_dir][:]
    multiplicitychargedpions_7GeV_lerrs=hdf[multiplicitychargedpions_lerrs_dir][:]

    multiplicitychargedkaons_7GeV_vals = hdf[multiplicitychargedkaons_vals_dir][:]
    multiplicitychargedkaons_7GeV_uerrs=hdf[multiplicitychargedkaons_uerrs_dir][:]
    multiplicitychargedkaons_7GeV_lerrs=hdf[multiplicitychargedkaons_lerrs_dir][:]

    multiplicityprotons_7GeV_vals = hdf[multiplicityprotons_vals_dir][:]
    multiplicityprotons_7GeV_uerrs=hdf[multiplicityprotons_uerrs_dir][:]
    multiplicityprotons_7GeV_lerrs=hdf[multiplicityprotons_lerrs_dir][:]
    centrality = hdf[centrality_dir][:]



multiplicitychargedpions_27GeV_symerr = 0.5 * (multiplicitychargedpions_27GeV_uerrs+multiplicitychargedpions_27GeV_lerrs)
multiplicitychargedkaons_27GeV_symerr = 0.5 * (multiplicitychargedkaons_27GeV_lerrs+multiplicitychargedkaons_27GeV_uerrs)
multiplicityprotons_27GeV_symerr = 0.5 * (multiplicityprotons_27GeV_uerrs+multiplicityprotons_27GeV_lerrs)

multiplicitychargedpions_19GeV_symerr = 0.5 * (multiplicitychargedpions_19GeV_uerrs+multiplicitychargedpions_19GeV_lerrs)
multiplicitychargedkaons_19GeV_symerr = 0.5 * (multiplicitychargedkaons_19GeV_lerrs+multiplicitychargedkaons_19GeV_uerrs)
multiplicityprotons_19GeV_symerr = 0.5 * (multiplicityprotons_19GeV_uerrs+multiplicityprotons_19GeV_lerrs)

multiplicitychargedpions_7GeV_symerr = 0.5 * (multiplicitychargedpions_7GeV_uerrs+multiplicitychargedpions_7GeV_lerrs)
multiplicitychargedkaons_7GeV_symerr = 0.5 * (multiplicitychargedkaons_7GeV_lerrs+multiplicitychargedkaons_7GeV_uerrs)
multiplicityprotons_7GeV_symerr = 0.5 * (multiplicityprotons_7GeV_uerrs+multiplicityprotons_7GeV_lerrs)
```

```python
Data_7GeV = np.loadtxt("/Users/christianschmidt/Downloads/Kp_proton_clean_7.7GeV.txt",skiprows=1)
Data_19GeV = np.loadtxt("/Users/christianschmidt/Downloads/Kp_proton_clean_19.6GeV.txt",skiprows=1)
Data_27GeV = np.loadtxt("/Users/christianschmidt/Downloads/Kp_proton_clean_27GeV.txt",skiprows=1)

Data_centrality = [2.5,7.5,15,25,35,45,55,65,75]

kplusoverpiplus_7GeV_data = Data_7GeV[:,1]
kplusoverpiplus_7GeV_staterr = Data_7GeV[:,2]
kplusoverpiplus_7GeV_syserr = Data_7GeV[:,3]
kplusoverpiplus_7GeV_quaderr = np.sqrt(kplusoverpiplus_7GeV_staterr**2+kplusoverpiplus_7GeV_syserr**2)
protonoverpiplus_7GeV_data = Data_7GeV[:,4]
protonoverpiplus_7GeV_staterr = Data_7GeV[:,5]
protonoverpiplus_7GeV_syserr = Data_7GeV[:,6]
protonoverpiplus_7GeV_quaderr = np.sqrt(protonoverpiplus_7GeV_staterr**2+protonoverpiplus_7GeV_syserr**2)

kplusoverpiplus_19GeV_data = Data_19GeV[:,1]
kplusoverpiplus_19GeV_staterr = Data_19GeV[:,2]
kplusoverpiplus_19GeV_syserr = Data_19GeV[:,3]
kplusoverpiplus_19GeV_quaderr = np.sqrt(kplusoverpiplus_19GeV_staterr**2+kplusoverpiplus_19GeV_syserr**2)
protonoverpiplus_19GeV_data = Data_19GeV[:,4]
protonoverpiplus_19GeV_staterr = Data_19GeV[:,5]
protonoverpiplus_19GeV_syserr = Data_19GeV[:,6]
protonoverpiplus_19GeV_quaderr = np.sqrt(protonoverpiplus_19GeV_staterr**2+protonoverpiplus_19GeV_syserr**2)

kplusoverpiplus_27GeV_data = Data_27GeV[:,1]
kplusoverpiplus_27GeV_staterr = Data_27GeV[:,2]
kplusoverpiplus_27GeV_syserr = Data_27GeV[:,3]
kplusoverpiplus_27GeV_quaderr = np.sqrt(kplusoverpiplus_27GeV_staterr**2+kplusoverpiplus_27GeV_syserr**2)
protonoverpiplus_27GeV_data = Data_27GeV[:,4]
protonoverpiplus_27GeV_staterr = Data_27GeV[:,5]
protonoverpiplus_27GeV_syserr = Data_27GeV[:,6]
protonoverpiplus_27GeV_quaderr = np.sqrt(protonoverpiplus_27GeV_staterr**2+protonoverpiplus_27GeV_syserr**2)
```

```python
koverpi_27GeV_trajectum_vals = multiplicitychargedkaons_27GeV_vals/multiplicitychargedpions_27GeV_vals
koverpi_27GeV_trajectum_err = np.sqrt((multiplicitychargedkaons_27GeV_symerr/multiplicitychargedpions_27GeV_vals)**2+(multiplicitychargedkaons_27GeV_vals*multiplicitychargedpions_27GeV_symerr/multiplicitychargedpions_27GeV_vals**2)**2)
poverpi_27GeV_trajectum_vals = multiplicityprotons_27GeV_vals/multiplicitychargedpions_27GeV_vals
poverpi_27GeV_trajectum_err = np.sqrt((multiplicityprotons_27GeV_symerr/multiplicitychargedpions_27GeV_vals)**2+(multiplicityprotons_27GeV_vals*multiplicitychargedpions_27GeV_symerr/multiplicitychargedpions_27GeV_vals**2)**2)

koverpi_19GeV_trajectum_vals = multiplicitychargedkaons_19GeV_vals/multiplicitychargedpions_19GeV_vals
koverpi_19GeV_trajectum_err = np.sqrt((multiplicitychargedkaons_19GeV_symerr/multiplicitychargedpions_19GeV_vals)**2+(multiplicitychargedkaons_19GeV_vals*multiplicitychargedpions_19GeV_symerr/multiplicitychargedpions_19GeV_vals**2)**2)
poverpi_19GeV_trajectum_vals = multiplicityprotons_19GeV_vals/multiplicitychargedpions_19GeV_vals
poverpi_19GeV_trajectum_err = np.sqrt((multiplicityprotons_19GeV_symerr/multiplicitychargedpions_19GeV_vals)**2+(multiplicityprotons_19GeV_vals*multiplicitychargedpions_19GeV_symerr/multiplicitychargedpions_19GeV_vals**2)**2)

koverpi_7GeV_trajectum_vals = multiplicitychargedkaons_7GeV_vals/multiplicitychargedpions_7GeV_vals
koverpi_7GeV_trajectum_err = np.sqrt((multiplicitychargedkaons_7GeV_symerr/multiplicitychargedpions_7GeV_vals)**2+(multiplicitychargedkaons_7GeV_vals*multiplicitychargedpions_7GeV_symerr/multiplicitychargedpions_7GeV_vals**2)**2)
poverpi_7GeV_trajectum_vals = multiplicityprotons_7GeV_vals/multiplicitychargedpions_7GeV_vals
poverpi_7GeV_trajectum_err = np.sqrt((multiplicityprotons_7GeV_symerr/multiplicitychargedpions_7GeV_vals)**2+(multiplicityprotons_7GeV_vals*multiplicitychargedpions_7GeV_symerr/multiplicitychargedpions_7GeV_vals**2)**2)
```

```python
plt.rcParams.update({"text.usetex": False, "mathtext.fontset": "cm", 
                     "font.family": "serif", "font.size": 12})
plt.figure(figsize=(8, 6), dpi=200)

plt.errorbar(centrality.ravel(),koverpi_27GeV_trajectum_vals.ravel(),yerr=koverpi_27GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="blue",label=r"$\frac{\kappa^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=27$ GeV")
plt.errorbar(Data_centrality,kplusoverpiplus_27GeV_data,yerr=kplusoverpiplus_27GeV_quaderr,fmt="^",capsize=5,color="blue",label=r"$\frac{\kappa^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=27$ GeV")

plt.errorbar(centrality.ravel(),koverpi_19GeV_trajectum_vals.ravel(),yerr=koverpi_19GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="red",label=r"$\frac{\kappa^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=19.6$ GeV")
plt.errorbar(Data_centrality,kplusoverpiplus_19GeV_data,yerr=kplusoverpiplus_19GeV_quaderr,fmt="^",capsize=5,color="red",label=r"$\frac{\kappa^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=19.6$ GeV")

plt.errorbar(centrality.ravel(),koverpi_7GeV_trajectum_vals.ravel(),yerr=koverpi_7GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="green",label=r"$\frac{\kappa^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=7.7$ GeV")
plt.errorbar(Data_centrality,kplusoverpiplus_7GeV_data,yerr=kplusoverpiplus_7GeV_quaderr,fmt="^",capsize=5,color="green",label=r"$\frac{\kappa^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=7.7$ GeV")


plt.grid()
plt.xlabel("Centrality (%)")
plt.ylabel("Particle ratio")
plt.xlim([0,80])
plt.ylim([0.085,0.3])
plt.legend(prop={'size': '8'},loc="lower left")
plt.show()

plt.rcParams.update({"text.usetex": False, "mathtext.fontset": "cm", 
                     "font.family": "serif", "font.size": 12})
plt.figure(figsize=(8, 6), dpi=200)

plt.errorbar(centrality.ravel(),poverpi_27GeV_trajectum_vals.ravel(),yerr=poverpi_27GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="blue",label=r"$\frac{p^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=27$ GeV")
plt.errorbar(Data_centrality,protonoverpiplus_27GeV_data,yerr=protonoverpiplus_27GeV_quaderr,fmt="^",capsize=5,color="blue",label=r"$\frac{p^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=27$ GeV")

plt.errorbar(centrality.ravel(),poverpi_19GeV_trajectum_vals.ravel(),yerr=poverpi_19GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="red",label=r"$\frac{p^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=19.6$ GeV")
plt.errorbar(Data_centrality,protonoverpiplus_19GeV_data,yerr=protonoverpiplus_19GeV_quaderr,fmt="^",capsize=5,color="red",label=r"$\frac{p^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=19.6$ GeV")

plt.errorbar(centrality.ravel(),poverpi_7GeV_trajectum_vals.ravel(),yerr=poverpi_7GeV_trajectum_err.ravel(),fmt="o",capsize=5,color="green",label=r"$\frac{p^\pm}{\pi^\pm}$, Trajectum, $\sqrt{s_{NN}}=7.7$ GeV")
plt.errorbar(Data_centrality,protonoverpiplus_7GeV_data,yerr=protonoverpiplus_7GeV_quaderr,fmt="^",capsize=5,color="green",label=r"$\frac{p^+}{\pi^+}$, arXiv:1701.07065, $\sqrt{s_{NN}}=7.7$ GeV")

plt.grid()
plt.xlabel("Centrality (%)")
plt.ylabel("Particle ratio")
plt.xlim([0,80])
plt.ylim([0,0.8])
plt.legend(prop={'size': '8'},loc="upper right")
plt.show()
```
