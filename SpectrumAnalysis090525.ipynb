```python
#27 GeV from data

with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-pi-_27_GeV.yaml", "r") as f:
    data_27GeV_piminus = yaml.safe_load(f)
with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-pi-_27_GeV.yaml", "r") as f:
    data_27GeV_piplus = yaml.safe_load(f)
with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-k-_27_GeV.yaml", "r") as f:
    data_27GeV_kminus = yaml.safe_load(f)
with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-k+_27_GeV.yaml", "r") as f:
    data_27GeV_kplus = yaml.safe_load(f)
with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-pbar_27_GeV.yaml", "r") as f:
    data_27GeV_pbar = yaml.safe_load(f)
with open("/Users/christianschmidt/Downloads/27GeV_Spectra/HEPData-ins1510593-v1-proton_27_GeV.yaml", "r") as f:
    data_27GeV_proton = yaml.safe_load(f)

dv = data_27GeV_piminus["dependent_variables"][0]  # first block
values = dv["values"]
piminus_27GeV_vals = np.array([v["value"] for v in values])
piminus_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
piminus_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_piminus["independent_variables"][0]
bins = iv["values"]
pt_low_piminus_27GeV = np.array([b["low"] for b in bins])
pt_high_piminus_27GeV = np.array([b["high"] for b in bins])
pt_centers_piminus_27GeV = 0.5 * (pt_low_piminus_27GeV + pt_high_piminus_27GeV)
pt_widths_piminus_27GeV  = pt_high_piminus_27GeV - pt_low_piminus_27GeV



dv = data_27GeV_piplus["dependent_variables"][0]  # first block
values = dv["values"]
piplus_27GeV_vals = np.array([v["value"] for v in values])
piplus_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
piplus_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_piplus["independent_variables"][0]
bins = iv["values"]
pt_low_piplus_27GeV = np.array([b["low"] for b in bins])
pt_high_piplus_27GeV = np.array([b["high"] for b in bins])
pt_centers_piplus_27GeV = 0.5 * (pt_low_piminus_27GeV + pt_high_piminus_27GeV)
pt_widths_piplus_27GeV  = pt_high_piminus_27GeV - pt_low_piminus_27GeV

dv = data_27GeV_kminus["dependent_variables"][0]  # first block
values = dv["values"]
kminus_27GeV_vals = np.array([v["value"] for v in values])
kminus_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
kminus_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_kminus["independent_variables"][0]
bins = iv["values"]
pt_low_kminus_27GeV = np.array([b["low"] for b in bins])
pt_high_kminus_27GeV = np.array([b["high"] for b in bins])
pt_centers_kminus_27GeV = 0.5 * (pt_low_kminus_27GeV + pt_high_kminus_27GeV)
pt_widths_kminus_27GeV  = pt_high_kminus_27GeV - pt_low_kminus_27GeV

dv = data_27GeV_kplus["dependent_variables"][0]  # first block
values = dv["values"]
kplus_27GeV_vals = np.array([v["value"] for v in values])
kplus_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
kplus_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_kplus["independent_variables"][0]
bins = iv["values"]
pt_low_kplus_27GeV = np.array([b["low"] for b in bins])
pt_high_kplus_27GeV = np.array([b["high"] for b in bins])
pt_centers_kplus_27GeV = 0.5 * (pt_low_kplus_27GeV + pt_high_kplus_27GeV)
pt_widths_kplus_27GeV  = pt_high_kplus_27GeV - pt_low_kplus_27GeV

dv = data_27GeV_pbar["dependent_variables"][0]  # first block
values = dv["values"]
pbar_27GeV_vals = np.array([v["value"] for v in values])
pbar_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
pbar_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_pbar["independent_variables"][0]
bins = iv["values"]
pt_low_pbar_27GeV  = np.array([b["low"] for b in bins])
pt_high_pbar_27GeV = np.array([b["high"] for b in bins])
pt_centers_pbar_27GeV = 0.5 * (pt_low_pbar_27GeV + pt_high_pbar_27GeV)
pt_widths_pbar_27GeV  = pt_high_pbar_27GeV - pt_low_pbar_27GeV

dv = data_27GeV_proton["dependent_variables"][0]  # first block
values = dv["values"]
proton_27GeV_vals = np.array([v["value"] for v in values])
proton_27GeV_err_stat = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "stat")
                      for v in values])
proton_27GeV_err_sys = np.array([next(e["symerror"] for e in v["errors"] if e["label"] == "sys")
                     for v in values])
iv = data_27GeV_proton["independent_variables"][0]
bins = iv["values"]
pt_low_proton_27GeV  = np.array([b["low"] for b in bins])
pt_high_proton_27GeV = np.array([b["high"] for b in bins])
pt_centers_proton_27GeV = 0.5 * (pt_low_proton_27GeV + pt_high_proton_27GeV)
pt_widths_proton_27GeV  = pt_high_proton_27GeV - pt_low_proton_27GeV

species = {
    "piminus": (pt_centers_piminus_27GeV, piminus_27GeV_vals, piminus_27GeV_err_stat, piminus_27GeV_err_sys),
    "piplus":  (pt_centers_piplus_27GeV,  piplus_27GeV_vals,  piplus_27GeV_err_stat,  piplus_27GeV_err_sys),
    "kminus":  (pt_centers_kminus_27GeV,  kminus_27GeV_vals,  kminus_27GeV_err_stat,  kminus_27GeV_err_sys),
    "kplus":   (pt_centers_kplus_27GeV,   kplus_27GeV_vals,   kplus_27GeV_err_stat,   kplus_27GeV_err_sys),
    "pbar":    (pt_centers_pbar_27GeV,    pbar_27GeV_vals,    pbar_27GeV_err_stat,    pbar_27GeV_err_sys),
    "proton":  (pt_centers_proton_27GeV,  proton_27GeV_vals,  proton_27GeV_err_stat,  proton_27GeV_err_sys),
}

# 1) find common centers by rounding
decimals = 8
sets = []
for name, (centers, vals, est_stat, est_sys) in species.items():
    rounded = np.round(centers, decimals)
    sets.append(set(rounded))
common_set = sets[0].copy()
for s in sets[1:]:
    common_set &= s

common_centers_27GeV = np.array(sorted(common_set))

if common_centers_27GeV.size == 0:
    raise RuntimeError("No common bin centers found among species. Try reducing rounding precision.")

# 2) build aligned arrays (for each species, pick values for common centers)
# We'll use rounded keys and dictionaries for safe mapping
def build_map(centers, values, decimals=decimals):
    rounded = np.round(centers, decimals)
    return {c: v for c, v in zip(rounded, values)}

vals_maps = {}
stat_maps = {}
sys_maps = {}
for name, (centers, vals, est_stat, est_sys) in species.items():
    vals_maps[name] = build_map(centers, vals)
    stat_maps[name] = build_map(centers, est_stat)
    sys_maps[name]  = build_map(centers, est_sys)

# Now create aligned arrays (shape: n_common)
n = common_centers_27GeV.size
aligned_vals = np.zeros((len(species), n), dtype=float)
aligned_stat_err = np.zeros_like(aligned_vals)
aligned_sys_err  = np.zeros_like(aligned_vals)

for i, name in enumerate(species.keys()):
    vm = vals_maps[name]
    sm = stat_maps[name]
    sx = sys_maps[name]
    for j, c in enumerate(common_centers_27GeV):
        aligned_vals[i, j] = vm[c]
        aligned_stat_err[i, j] = sm[c]
        aligned_sys_err[i, j]  = sx[c]

# 3) sum values across species for each common bin
total_vals_common_27GeV = np.sum(aligned_vals, axis=0) * common_centers_27GeV * 2 * np.pi

# 4) combine errors in quadrature across species
#    if stat/sys are independent across species, variances add:
combined_stat_err = np.sqrt(np.sum(aligned_stat_err**2, axis=0))
combined_sys_err  = np.sqrt(np.sum(aligned_sys_err**2, axis=0))

# If you want a single total error per bin (stat+sys combined), do:
total_err_common_27GeV = np.sqrt(combined_stat_err**2 + combined_sys_err**2) * common_centers_27GeV * 2 * np.pi

# 5) plot
plt.errorbar(common_centers_27GeV, total_vals_common_27GeV, yerr=total_err_common_27GeV,
             fmt='o', capsize=5, label='sum of species (27 GeV)')
plt.xlabel("pT (GeV/c)")
plt.ylabel("Yield")
plt.legend()
plt.show()
```
